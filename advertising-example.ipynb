```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("Advertising.csv",index_col=0 )
```

```python
df.describe()
```

```python
df.head()
```

```python
x_1 = df["TV"]
```

```python
x_2 = df["radio"]
```

```python
x_3 = df["newspaper"]
```

```python
y = df["sales"]
```

```python
linear_regressor = LinearRegression()
```

```python
y_r = y.values.reshape(-1,1)
```

```python
x1_r = x_1.values.reshape(-1,1)
x2_r = x_2.values.reshape(-1,1)
x3_r = x_3.values.reshape(-1,1)
```

```python
plt.scatter(x_1,y)
linear_regressor.fit(x1_r,y)
yPred_1 = linear_regressor.predict(x1_r)
plt.plot(x_1,yPred_1,"y")

plt.xlabel("TV")
plt.ylabel("Sales")
```

```python
plt.scatter(x_2,y)
linear_regressor.fit(x2_r,y)
yPred_2 = linear_regressor.predict(x2_r)
plt.plot(x_2,yPred_2,"y")

plt.xlabel("radio")
plt.ylabel("Sales")
```

```python
plt.scatter(x_3,y)
linear_regressor.fit(x3_r,y)
yPred_3 = linear_regressor.predict(x3_r)
plt.plot(x_3,yPred_3,"y")

plt.xlabel("newspaper")
plt.ylabel("Sales")
```
